```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.animation as animation
```

```python
fig, ax = plt.subplots()
xs = list(range(0, 10))
ys = list(range(0, 10))

scat = ax.scatter(xs[0], ys[0], c="b", s=5)
ax.set(xlim=[-1, 11], ylim=[-1, 11])

def init():
    scat.set_offsets([[xs[0], ys[0]]])
    return scat,
    
def update(frame):
    scat.set_offsets([[xs[frame], ys[frame]]])
    return scat,


ani = animation.FuncAnimation(fig=fig, func=update, init_func=init, frames=10)
ani.save(filename="anim.gif", writer="pillow")
# plt.show()
```

```python
x = []
y = []

figure, ax = plt.subplots()

# Setting limits for x and y axis
ax.set_xlim(0, 100)
ax.set_ylim(0, 12)

# Since plotting a single graph
line,  = ax.plot(0, 0) 

def animation_function(i):
    x.append(i * 15)
    y.append(i)
    # x = i * 15
    # y = i

    line.set_xdata(x)
    line.set_ydata(y)
    return line,

ani = animation.FuncAnimation(figure,
                          func = animation_function,
                          frames = np.arange(0, 10, 0.1), 
                          interval = 10)
plt.show()
```

```python
# Define your x and y points
x_points = [1, 2, 3, 4, 5]  # Example x coordinates
y_points = [1, 4, 9, 16, 25]  # Example y coordinates (squared values)

# Set up the figure and axis
fig, ax = plt.subplots()
ax.set_xlim(min(x_points) - 1, max(x_points) + 1)
ax.set_ylim(min(y_points) - 1, max(y_points) + 1)

# Initialize the point plot, which will be updated in each frame
point, = ax.plot([], [], 'bo', markersize=8)  # 'bo' means blue circle marker

# Function to initialize the background of the animation
def init():
    point.set_data([], [])
    return point,

# Function to update the animation with each frame
def update(frame):
    # Set the point to the current frame's x and y data
    point.set_data(x_points[frame], y_points[frame])
    return point,

# Create the animation
ani = animation.FuncAnimation(
    fig,       # The figure object
    update,    # Update function
    frames=len(x_points),  # Number of frames (one per point)
    init_func=init,  # Initialization function
    blit=True   # Use blitting for faster animation
)

# Show the animation
plt.show()
```
